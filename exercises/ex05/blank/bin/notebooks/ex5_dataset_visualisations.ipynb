```python
import pandas as pd
from nlpwdlfw.datasets import SyntheticBinaryClassificationDataset

dataset = SyntheticBinaryClassificationDataset()
training_examples = dataset.get_training_examples()

# convert examples to a numpy array
raw_data = []
for example in training_examples:
    row = []
    row.extend(example.feature_vector)
    row.append(example.label)
    raw_data.append(row)

df = pd.DataFrame(raw_data, columns=["x1", "x2", "label"])
df
```

```python
import seaborn as sns

df_pos = df.loc[df["label"] == 1]
df_neg = df.loc[df["label"] == 0]
sns.scatterplot(data=df_pos, x="x1", y="x2", )
sns.scatterplot(data=df_neg, x="x1", y="x2", )
```
